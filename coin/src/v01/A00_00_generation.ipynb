```python
import pandas as pd
import numpy as np
import yfinance as yf
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

class CryptoFeatureBuilder:
    def __init__(self):
        self.periods = [1, 3, 7, 14, 30, 90]  # κΈ°λ³Έ κΈ°κ°„λ“¤
        
    def fetch_crypto_data(self, symbols= major_cryptos, start_date='2015-01-01', end_date='2025-09-27'):
        """μ•Όν›„ νμ΄λ‚Έμ¤μ—μ„ μ•”νΈν™”ν λ°μ΄ν„° κ°€μ Έμ¤κΈ°"""
        all_data = []
        for symbol in symbols:
            try:
                ticker = yf.Ticker(symbol)
                data = ticker.history(start=start_date, end=end_date, interval='1d')
                data['Symbol'] = symbol.replace('-USD', '')
                data.reset_index(inplace=True)
                all_data.append(data)
                print(f"β… {symbol} λ°μ΄ν„° μμ§‘ μ™„λ£: {len(data)} μΌ ({data['Date'].min().date()} ~ {data['Date'].max().date()})")
            except Exception as e:
                print(f"β {symbol} λ°μ΄ν„° μμ§‘ μ‹¤ν¨: {e}")
        
        if all_data:
            combined_data = pd.concat(all_data, ignore_index=True)
            return combined_data.sort_values(['Symbol', 'Date']).reset_index(drop=True)
        return pd.DataFrame()
    
    def calculate_basic_indicators(self, df):
        """κΈ°λ³Έ μ§€ν‘λ“¤ κ³„μ‚°"""
        # κ°€κ²© κ΄€λ ¨ λ¦¬ν„΄
        df['Close_return'] = df.groupby('Symbol')['Close'].pct_change()
        df['High_return'] = df.groupby('Symbol')['High'].pct_change()
        df['Low_return'] = df.groupby('Symbol')['Low'].pct_change()
        df['Open_return'] = df.groupby('Symbol')['Open'].pct_change()
        
        # λ΅κ·Έ λ¦¬ν„΄ - transform μ‚¬μ©
        df['Close_log_return'] = df.groupby('Symbol')['Close'].transform(lambda x: np.log(x/x.shift(1)))
        df['High_log_return'] = df.groupby('Symbol')['High'].transform(lambda x: np.log(x/x.shift(1)))
        df['Low_log_return'] = df.groupby('Symbol')['Low'].transform(lambda x: np.log(x/x.shift(1)))
        df['Open_log_return'] = df.groupby('Symbol')['Open'].transform(lambda x: np.log(x/x.shift(1)))
        
        # λ³Όλ¥¨ μ •κ·ν™” - transform μ‚¬μ©
        df['Volume_normalized'] = df.groupby('Symbol')['Volume'].transform(
            lambda x: (x - x.rolling(30).mean()) / x.rolling(30).std()
        )
        df['Volume_log_ratio'] = df.groupby('Symbol')['Volume'].transform(lambda x: np.log(x/x.shift(1)))
        
        return df
    
    def calculate_ma_indicators(self, df):
        """μ΄λ™ν‰κ·  μ§€ν‘λ“¤"""
        ma_periods = [3, 7, 14, 30]
        
        for period in ma_periods:
            # λ‹¨μ μ΄λ™ν‰κ· 
            df[f'MA_{period}'] = df.groupby('Symbol')['Close'].transform(lambda x: x.rolling(period).mean())
            df[f'MA_{period}_ratio'] = df['Close'] / df[f'MA_{period}'] - 1
            df[f'MA_{period}_distance'] = (df['Close'] - df[f'MA_{period}']) / df['Close']
            
            # μ§€μ μ΄λ™ν‰κ· 
            df[f'EMA_{period}'] = df.groupby('Symbol')['Close'].transform(lambda x: x.ewm(span=period).mean())
            df[f'EMA_{period}_ratio'] = df['Close'] / df[f'EMA_{period}'] - 1
        
        # MA ν¬λ΅μ¤ μ‹ νΈ
        if 'MA_3' in df.columns and 'MA_30' in df.columns:
            df['MA_Cross_Signal'] = np.where(df['MA_3'] > df['MA_30'], 1, 0)
            df['MA_Cross_Signal_ratio'] = df['MA_3'] / df['MA_30'] - 1
            df['MA_Cross_Signal_distance'] = (df['MA_3'] - df['MA_30']) / df['Close']
        
        return df
    
    def calculate_bollinger_bands(self, df, period=20):
        """λ³Όλ¦°μ € λ°΄λ“"""
        df[f'BB_middle'] = df.groupby('Symbol')['Close'].transform(lambda x: x.rolling(period).mean())
        bb_std = df.groupby('Symbol')['Close'].transform(lambda x: x.rolling(period).std())
        df[f'BB_upper'] = df[f'BB_middle'] + 2 * bb_std
        df[f'BB_lower'] = df[f'BB_middle'] - 2 * bb_std
        
        df['BB_relative_position'] = (df['Close'] - df[f'BB_lower']) / (df[f'BB_upper'] - df[f'BB_lower'])
        df['BB_upper_distance'] = (df[f'BB_upper'] - df['Close']) / df['Close']
        df['BB_lower_distance'] = (df['Close'] - df[f'BB_lower']) / df['Close']
        
        return df
    
    def calculate_rsi_variations(self, df):
        """λ‹¤μ–‘ν• RSI κΈ°κ°„κ³Ό μ‹ νΈ + λ‹¤μ΄λ²„μ „μ¤"""
        rsi_periods = [3, 7, 14, 30]
        
        for period in rsi_periods:
            delta = df.groupby('Symbol')['Close'].diff()
            gain = delta.where(delta > 0, 0).groupby(df['Symbol']).transform(lambda x: x.rolling(period).mean())
            loss = (-delta.where(delta < 0, 0)).groupby(df['Symbol']).transform(lambda x: x.rolling(period).mean())
            rs = gain / loss
            df[f'RSI_{period}'] = 100 - (100 / (1 + rs))
            
            # RSI μ‹ νΈλ“¤
            df[f'RSI_{period}_overbought'] = (df[f'RSI_{period}'] > 70).astype(int)
            df[f'RSI_{period}_oversold'] = (df[f'RSI_{period}'] < 30).astype(int)
            df[f'RSI_{period}_normalized'] = (df[f'RSI_{period}'] - 50) / 50
        
        # RSI λ‹¤μ΄λ²„μ „μ¤ κ³„μ‚° (14μΌ, 7μΌ)
        for period in [7, 14]:
            if f'RSI_{period}' in df.columns:
                # μƒμΉ λ‹¤μ΄λ²„μ „μ¤: κ°€κ²©μ€ ν•λ½ν•μ§€λ§ RSIλ” μƒμΉ
                price_lower = (df['Close'] < df['Close'].shift(5)) & (df['Close'].shift(5) < df['Close'].shift(10))
                rsi_higher = (df[f'RSI_{period}'] > df[f'RSI_{period}'].shift(5)) & (df[f'RSI_{period}'].shift(5) > df[f'RSI_{period}'].shift(10))
                df[f'RSI_{period}_bullish_divergence'] = (price_lower & rsi_higher).astype(int)
                
                # ν•λ½ λ‹¤μ΄λ²„μ „μ¤: κ°€κ²©μ€ μƒμΉν•μ§€λ§ RSIλ” ν•λ½
                price_higher = (df['Close'] > df['Close'].shift(5)) & (df['Close'].shift(5) > df['Close'].shift(10))
                rsi_lower = (df[f'RSI_{period}'] < df[f'RSI_{period}'].shift(5)) & (df[f'RSI_{period}'].shift(5) < df[f'RSI_{period}'].shift(10))
                df[f'RSI_{period}_bearish_divergence'] = (price_higher & rsi_lower).astype(int)
        
        return df
    
    def calculate_macd_variations(self, df):
        """MACD μ§€ν‘λ“¤"""
        configs = [(7, 14), (14, 30)]
        
        for fast, slow in configs:
            ema_fast = df.groupby('Symbol')['Close'].transform(lambda x: x.ewm(span=fast).mean())
            ema_slow = df.groupby('Symbol')['Close'].transform(lambda x: x.ewm(span=slow).mean())
            macd_line = ema_fast - ema_slow
            signal_line = macd_line.groupby(df['Symbol']).transform(lambda x: x.ewm(span=9).mean())
            
            df[f'MACD_{fast}_{slow}'] = macd_line
            df[f'MACD_{fast}_{slow}_signal'] = signal_line
            df[f'MACD_{fast}_{slow}_above_signal'] = (macd_line > signal_line).astype(int)
            df[f'MACD_{fast}_{slow}_signal_distance'] = (macd_line - signal_line) / df['Close']
        
        return df
    
    def calculate_stochastic_variations(self, df):
        """λ‹¤μ–‘ν• μ¤ν† μΊμ¤ν‹± μ§€ν‘λ“¤ - Top20μ— ν¬ν•¨λ κ²ƒλ“¤"""
        stoch_configs = [(1, 3), (3, 3), (4, 3), (5, 3), (6, 3), (14, 3)]
        
        for k_period, d_period in stoch_configs:
            low_k = df.groupby('Symbol')['Low'].transform(lambda x: x.rolling(k_period).min())
            high_k = df.groupby('Symbol')['High'].transform(lambda x: x.rolling(k_period).max())
            range_k = high_k - low_k
            
            df[f'Stoch_K_{k_period}'] = np.where(
                range_k > 0,
                100 * ((df['Close'] - low_k) / range_k),
                50
            )
            
            df[f'Stoch_D_{k_period}'] = df.groupby('Symbol')[f'Stoch_K_{k_period}'].transform(lambda x: x.rolling(d_period).mean())
            
            # Top20μ— ν¬ν•¨λ μ‹ νΈλ“¤
            df[f'Stoch_{k_period}_K_above_D'] = (df[f'Stoch_K_{k_period}'] > df[f'Stoch_D_{k_period}']).astype(int)
            df[f'Stoch_{k_period}_overbought'] = (df[f'Stoch_K_{k_period}'] > 80).astype(int)
            df[f'Stoch_{k_period}_oversold'] = (df[f'Stoch_K_{k_period}'] < 20).astype(int)
        
        # Stoch_K_3 (Top20μ—μ„ ν•„μ”)
        df['Stoch_K_3'] = df['Stoch_K_3']
            
        return df
    
    def calculate_cci_variations(self, df):
        """CCI μ§€ν‘λ“¤ - Top20μ— ν¬ν•¨λ κ²ƒλ“¤"""
        cci_periods = [3, 4, 7, 14]
        
        for period in cci_periods:
            tp = (df['High'] + df['Low'] + df['Close']) / 3
            cci_ma = tp.groupby(df['Symbol']).transform(lambda x: x.rolling(period).mean())
            cci_mad = tp.groupby(df['Symbol']).transform(
                lambda x: x.rolling(period).apply(lambda y: np.mean(np.abs(y - np.mean(y))))
            )
            
            df[f'CCI_{period}'] = np.where(
                cci_mad > 0,
                (tp - cci_ma) / (0.015 * cci_mad),
                0
            )
            
            # Top20μ— ν¬ν•¨λ μ‹ νΈλ“¤
            df[f'CCI_{period}_overbought'] = (df[f'CCI_{period}'] > 100).astype(int)
            df[f'CCI_{period}_oversold'] = (df[f'CCI_{period}'] < -100).astype(int)
        
        return df
    
    def calculate_smi(self, df):
        """Stochastic Momentum Index κ³„μ‚° - νμΈμ¤ν¬λ¦½νΈ μ •ν™•ν κµ¬ν„"""
        # νμΈμ¤ν¬λ¦½νΈ νλΌλ―Έν„° κΈ°μ¤€
        k_length = 10   # Percent K Length (a)
        d_length = 3    # Percent D Length (b)
        ema_length = 10 # EMA Signal Length

        # Range Calculation - νμΈμ¤ν¬λ¦½νΈμ™€ λ™μΌ
        ll = df.groupby('Symbol')['Low'].transform(lambda x: x.rolling(k_length).min())
        hh = df.groupby('Symbol')['High'].transform(lambda x: x.rolling(k_length).max())
        diff = hh - ll
        rdiff = df['Close'] - (hh + ll) / 2

        # νμΈμ¤ν¬λ¦½νΈ: avgrel = ta.ema(ta.ema(rdiff, b), b)
        # adjust=False μ¶”κ°€
        avgrel_step1 = rdiff.groupby(df['Symbol']).transform(lambda x: x.ewm(span=d_length, adjust=False).mean())
        avgrel = avgrel_step1.groupby(df['Symbol']).transform(lambda x: x.ewm(span=d_length, adjust=False).mean())

        # νμΈμ¤ν¬λ¦½νΈ: avgdiff = ta.ema(ta.ema(diff, b), b)
        # adjust=False μ¶”κ°€
        avgdiff_step1 = diff.groupby(df['Symbol']).transform(lambda x: x.ewm(span=d_length, adjust=False).mean())
        avgdiff = avgdiff_step1.groupby(df['Symbol']).transform(lambda x: x.ewm(span=d_length, adjust=False).mean())

        # SMI κ³„μ‚°: SMI = avgdiff != 0 ? avgrel / (avgdiff / 2) * 100 : 0
        df['SMI'] = np.where(avgdiff != 0, (avgrel / (avgdiff / 2)) * 100, 0)

        # SMI Signal: SMIsignal = ta.ema(SMI, b)
        # adjust=False μ¶”κ°€
        df['SMI_signal'] = df.groupby('Symbol')['SMI'].transform(lambda x: x.ewm(span=d_length, adjust=False).mean())

        # EMA Signal: emasignal = ta.ema(SMI, 10)
        # adjust=False μ¶”κ°€
        df['SMI_ema'] = df.groupby('Symbol')['SMI'].transform(lambda x: x.ewm(span=ema_length, adjust=False).mean())

        # SMI μ‹ νΈλ“¤
        df['SMI_overbought_40'] = (df['SMI_signal'] > 40).astype(int)
        df['SMI_oversold_40'] = (df['SMI_signal'] < -40).astype(int)
        df['SMI_overbought_50'] = (df['SMI_signal'] > 50).astype(int)
        df['SMI_oversold_50'] = (df['SMI_signal'] < -50).astype(int)

        # μ •κ·ν™”
        df['SMI_normalized'] = np.clip(df['SMI'] / 100, -1, 1)

        return df
        
    def calculate_volatility(self, df):
        """λ³€λ™μ„± μ§€ν‘λ“¤"""
        periods = [7, 14, 30]
        
        for period in periods:
            df[f'Volatility_{period}d'] = df.groupby('Symbol')['Close_return'].transform(lambda x: x.rolling(period).std()) * np.sqrt(365)
            
        # λ†’μ€ λ³€λ™μ„± μ²΄μ  (Top20μ— ν¬ν•¨)
        if 'Volatility_7d' in df.columns:
            df['high_volatility_regime'] = (
                df['Volatility_7d'] > df.groupby('Symbol')['Volatility_7d'].transform(lambda x: x.rolling(30).quantile(0.8))
            ).astype(int)
        
        # μ‹¤ν„ λ³€λ™μ„±
        df['realized_volatility_7'] = df.groupby('Symbol')['Close_log_return'].transform(lambda x: x.rolling(7).std()) * np.sqrt(365)
        df['realized_volatility_30'] = df.groupby('Symbol')['Close_log_return'].transform(lambda x: x.rolling(30).std()) * np.sqrt(365)
        
        return df
    
    def calculate_momentum_signals(self, df):
        """λ¨λ©ν…€ μ‹ νΈλ“¤"""
        # Return_7d κ³„μ‚°
        df['Return_7d'] = df.groupby('Symbol')['Close'].pct_change(7)
        
        # Momentum_Signal (Top20μ— ν¬ν•¨)
        df['Momentum_Signal'] = 1
        strong_momentum_up = (df['Return_7d'] > 0.1) & (df['Return_7d'].shift(1) <= 0.1)
        strong_momentum_down = (df['Return_7d'] < -0.1) & (df['Return_7d'].shift(1) >= -0.1)
        df.loc[strong_momentum_up, 'Momentum_Signal'] = 2
        df.loc[strong_momentum_down, 'Momentum_Signal'] = 0
        
        # Formula3_Signal (κ°€κ²© λ¨λ©ν…€ κΈ°λ°, Top20μ— ν¬ν•¨)
        df['Price_change_3d'] = df.groupby('Symbol')['Close'].pct_change(3)
        volume_ma = df.groupby('Symbol')['Volume'].transform(lambda x: x.rolling(7).mean())
        
        df['Formula3_Signal'] = 1
        formula3_up = (df['Price_change_3d'] > 0.05) & (df['Volume'] > volume_ma)
        formula3_down = (df['Price_change_3d'] < -0.05) & (df['Volume'] > volume_ma)
        df.loc[formula3_up, 'Formula3_Signal'] = 2
        df.loc[formula3_down, 'Formula3_Signal'] = 0
        
        return df
    
    def calculate_datetime_features(self, df):
        """λ‚ μ§/μ‹κ°„ νΉμ„±λ“¤"""
        df['Date'] = pd.to_datetime(df['Date'])
        df['year'] = df['Date'].dt.year
        df['month'] = df['Date'].dt.month
        df['quarter'] = df['Date'].dt.quarter
        df['day_of_week'] = df['Date'].dt.dayofweek
        
        # μ›”/λ¶„κΈ° μ‹μ‘/λ (Top20μ— ν¬ν•¨)
        df['is_month_start'] = df['Date'].dt.is_month_start.astype(int)
        df['is_month_end'] = df['Date'].dt.is_month_end.astype(int)
        df['is_quarter_start'] = df['Date'].dt.is_quarter_start.astype(int)
        df['is_quarter_end'] = df['Date'].dt.is_quarter_end.astype(int)
        
        # μν™ μΈμ½”λ”© (Top20μ— ν¬ν•¨)
        df['month_sin'] = np.sin(2 * np.pi * df['month'] / 12)
        df['month_cos'] = np.cos(2 * np.pi * df['month'] / 12)
        df['dow_sin'] = np.sin(2 * np.pi * df['day_of_week'] / 7)
        df['dow_cos'] = np.cos(2 * np.pi * df['day_of_week'] / 7)
        
        return df
    
    def calculate_future_returns(self, df):
        """λ―Έλ μμµλ¥  κ³„μ‚°"""
        future_periods = [1, 3, 7, 14, 30, 90]
        
        for period in future_periods:
            df[f'Future_Return_{period}d'] = df.groupby('Symbol')['Close'].shift(-period) / df['Close'] - 1
            
            # λΌλ²¨λ§ (μƒμΉ/ν•λ½)
            df[f'Future_Label_{period}d'] = (df[f'Future_Return_{period}d'] > 0).astype(int)
        
        return df
    
    def build_features(self, symbols=['BTC-USD', 'ETH-USD'], start_date='2022-01-01', end_date='2024-12-31'):
        """μ „μ²΄ νΉμ„± μƒμ„± νμ΄ν”„λΌμΈ"""
        print("π€ μ•”νΈν™”ν λ°μ΄ν„° μμ§‘ μ‹μ‘...")
        df = self.fetch_crypto_data(symbols, start_date, end_date)
        
        if df.empty:
            print("β λ°μ΄ν„° μμ§‘ μ‹¤ν¨")
            return df
        
        print("π“ κΈ°λ³Έ μ§€ν‘ κ³„μ‚° μ¤‘...")
        df = self.calculate_basic_indicators(df)
        
        print("π“ μ΄λ™ν‰κ·  μ§€ν‘ κ³„μ‚° μ¤‘...")
        df = self.calculate_ma_indicators(df)
        
        print("π“‰ λ³Όλ¦°μ € λ°΄λ“ κ³„μ‚° μ¤‘...")
        df = self.calculate_bollinger_bands(df)
        
        print("π“ RSI μ§€ν‘ κ³„μ‚° μ¤‘...")
        df = self.calculate_rsi_variations(df)
        
        print("π“ MACD μ§€ν‘ κ³„μ‚° μ¤‘...")
        df = self.calculate_macd_variations(df)
        
        print("π“ μ¤ν† μΊμ¤ν‹± μ§€ν‘ κ³„μ‚° μ¤‘...")
        df = self.calculate_stochastic_variations(df)
        
        print("π“‰ CCI μ§€ν‘ κ³„μ‚° μ¤‘...")
        df = self.calculate_cci_variations(df)
        
        print("π“ SMI μ§€ν‘ κ³„μ‚° μ¤‘...")
        df = self.calculate_smi(df)
        
        print("π“ λ³€λ™μ„± μ§€ν‘ κ³„μ‚° μ¤‘...")
        df = self.calculate_volatility(df)
        
        print("π€ λ¨λ©ν…€ μ‹ νΈ κ³„μ‚° μ¤‘...")
        df = self.calculate_momentum_signals(df)
        
        print("π“… λ‚ μ§/μ‹κ°„ νΉμ„± κ³„μ‚° μ¤‘...")
        df = self.calculate_datetime_features(df)
        
        print("π― λ―Έλ μμµλ¥  κ³„μ‚° μ¤‘...")
        df = self.calculate_future_returns(df)
        
        # NaN κ°’ μ •λ¦¬ - Future μ»¬λΌλ§ μ²λ¦¬
        print("π§Ή NaN κ°’ μ •λ¦¬ μ¤‘...")
        initial_rows = len(df)
        
        # Future μ»¬λΌλ“¤μ NaNμ„ νΉμ • κ°’μΌλ΅ μ±„μ°κΈ°
        future_cols = [col for col in df.columns if 'Future' in col]
        for col in future_cols:
            if 'Return' in col:
                df[col] = df[col].fillna(0)  # λ―Έλ μμµλ¥ μ€ 0μΌλ΅
            elif 'Label' in col:
                df[col] = df[col].fillna(1)  # λ―Έλ λΌλ²¨μ€ 1(Hold)λ΅
        
        # λ‹¤λ¥Έ κΈ°μ μ§€ν‘ NaNμ€ forward fill λλ” 0μΌλ΅
        df = df.fillna(method='ffill').fillna(0)
        
        final_rows = len(df)
        print(f"   μ΄κΈ° ν–‰μ: {initial_rows} β†’ μµμΆ… ν–‰μ: {final_rows} (λ³΄μ΅΄λ¨!)")
        
        # Top20 μ¤‘μ” νΉμ„±λ§ μ„ νƒ
        top20_features = [
            'Stoch_14_K_above_D', 'Stoch_6_K_above_D', 'CCI_4_overbought', 'Formula3_Signal',
            'Stoch_1_K_above_D', 'Stoch_4_K_above_D', 'is_quarter_start', 'high_volatility_regime',
            'is_month_end', 'year', 'Stoch_3_K_above_D', 'Stoch_K_3',
            'is_month_start', 'Momentum_Signal', 'day_of_week', 'Volatility_7d',
            'month_sin', 'CCI_3', 'dow_sin', 'Stoch_5_overbought'
        ]
        
        # κΈ°λ³Έ μ»¬λΌλ“¤ + Top20 + SMI + RSI λ‹¤μ΄λ²„μ „μ¤ + Future μ»¬λΌλ“¤λ§ μ„ νƒ
        essential_cols = ['Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Symbol']
        future_cols = [col for col in df.columns if 'Future' in col]
        available_top20 = [col for col in top20_features if col in df.columns]
        
        # SMI κ΄€λ ¨ μ»¬λΌλ“¤ μ¶”κ°€
        smi_cols = [col for col in df.columns if 'SMI' in col]
        
        # RSI λ‹¤μ΄λ²„μ „μ¤ μ»¬λΌλ“¤ μ¶”κ°€
        rsi_divergence_cols = [col for col in df.columns if 'divergence' in col]
        
        selected_cols = essential_cols + available_top20 + smi_cols + rsi_divergence_cols + future_cols
        df_selected = df[selected_cols].copy()
        
        print(f"β… νΉμ„± μƒμ„± μ™„λ£! λ°μ΄ν„° ν•νƒ: {df_selected.shape}")
        print(f"π“ μ„ νƒλ μ»¬λΌ μ: {len(df_selected.columns)}")
        print(f"π― Top20 μ¤‘ μ‚¬μ© κ°€λ¥ν• νΉμ„±: {len(available_top20)}/20")
        print(f"π“ SMI κ΄€λ ¨ νΉμ„±: {len(smi_cols)}κ°")
        print(f"π“ RSI λ‹¤μ΄λ²„μ „μ¤ νΉμ„±: {len(rsi_divergence_cols)}κ°")
        
        return df_selected
```
